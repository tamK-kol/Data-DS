Standard Deviation: This measures the dispersion of the property prices. A low standard deviation indicates that the prices are close to the mean, while a high standard deviation indicates that the prices are spread out over a wider range.

#### Inferential Statistics (6 marks):
● Formulate a hypothesis related to the dataset.

● Conduct a statistical test (e.g., t-test) to test the hypothesis.

● Provide conclusions based on the test results

Let’s try to perform inferential statistics on the dataset using Python. Inferential statistics are used to draw conclusions about a population based on a sample of data.

Formulate a hypothesis: A hypothesis is a statement that expresses a relationship between two variables. For example, we can formulate a hypothesis that the mean property price in Melbourne is different for different regions. The null hypothesis (H0) is that there is no difference in the mean price across regions, and the alternative hypothesis (H1) is that there is a difference.

Conduct a statistical test: A statistical test is a procedure that compares the observed data with the expected data under the null hypothesis. There are different types of tests depending on the data and the hypothesis. For example, we can use an ANOVA (analysis of variance) test to compare the mean price across multiple regions. Here’s how we can do it in Python using the scipy library:

```python
# Import pandas and scipy
import pandas as pd
from scipy import stats

# Load the data
df = pd.read_csv('/content/melb_data.csv')

# Conduct an ANOVA test
f_stat, p_value = stats.f_oneway(*[group['Price'].values for name, group in df.groupby('Regionname')])

# Print the test results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
```

Conclusions:

The ANOVA test results show that the p-value is 0.0, which is less than the common significance level of 0.05. This means that we can reject the null hypothesis and accept the alternative hypothesis that there is a difference in the mean property price across regions.

The F-statistic is 421.38, which is the ratio of the between-group variance to the within-group variance. A high F-statistic indicates that the variation between the groups is much larger than the variation within the groups, which supports the conclusion that the mean price is not the same for all regions.

#### Outlier Detection (5 marks):
● Identify and handle outliers in the dataset.

● Explain the methodology and reasoning behind your approach.

Outliers are data points that deviate significantly from the rest of the data. They can affect the accuracy and performance of machine learning models.

One way to identify outliers is to use boxplots, which show the distribution of the data and the quartiles. Any data point that lies beyond 1.5 times the interquartile range (IQR) from the median is considered an outlier.

Another way to identify outliers is to use z-scores, which measure how many standard deviations a data point is away from the mean. A common threshold for outliers is a z-score of 3 or more, or -3 or less.

One way to handle outliers is to remove them from the dataset, but this could result in loss of information and bias. Another way to handle outliers is to replace them with the mean or median of the data, but this could reduce the variability and introduce errors. A third way to handle outliers is to use robust methods that are less sensitive to outliers, such as median or mode for central tendency, or IQR or MAD for dispersion.


```python
# Create a boxplot of 'Price'
df['Price'].plot(kind='box', title='Price Boxplot')
plt.ylabel('Property Price')
plt.show()

# Calculate the lower and upper bounds for outliers
Q1 = df['Price'].quantile(0.25) # first quartile
Q3 = df['Price'].quantile(0.75) # third quartile
IQR = Q3 - Q1 # interquartile range
lower_bound = Q1 - 1.5 * IQR # lower bound for outliers
upper_bound = Q3 + 1.5 * IQR # upper bound for outliers

# Filter out the outliers from the dataset
df_no_outliers = df[(df['Price'] > lower_bound) & (df['Price'] < upper_bound)]

# Display the new dataset
df_no_outliers.head()
```

We chose to use boxplots to identify outliers in the ‘Price’ column, as they show the distribution and variation of the data and the quartiles. We considered any data point that lies beyond 1.5 times the interquartile range (IQR) from the median as an outlier, as this is a common rule of thumb. We then removed the outliers from the dataset, as this can reduce the noise and improve the accuracy of the data analysis and machine learning models. However, this approach also has some limitations, such as losing valuable information and reducing the sample size. Alternatively, we could have replaced the outliers with the mean or median of the data, or used robust methods that are less sensitive to outliers.
